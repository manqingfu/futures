```python
count=0
for i in range(len(data)-1):
    if abs(data.iloc[i+1]['con_closing']-data.iloc[i]['con_closing'])/data.iloc[i]['con_closing']>0.07:
        count+=1
        print(abs(data.iloc[i+1]['con_closing']-data.iloc[i]['con_closing'])/data.iloc[i]['con_closing'],data.iloc[i]['date'],data.iloc[i]['con_closing'])
count/len(data)
```

#### 期货交割日期

```python
import pandas as pd
data=pd.read_csv(r'data.csv')
# data
```

```python
settle_date=[]
count=0
for i in range(len(data)-1):
    if (data.iloc[i+1]['con_invent']>3*data.iloc[i]['con_invent']):
        settle_date.append(data.iloc[i]['date'])
        print(data.iloc[i]['date'],data.iloc[i-1]['con_closing'],data.iloc[i]['con_closing'],data.iloc[i+1]['con_closing'])
        count+=1
print(count)
```
